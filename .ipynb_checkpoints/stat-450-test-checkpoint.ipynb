```r
model <- lm(target_q_mean ~. , data = df)
summary(model)
```

```r
Best_Subset <-
    regsubsets(target_q_mean~.,
               data =df,
               nbest = 1,      # 1 best model for each number of predictors
               nvmax = NULL,    # NULL for no limit on number of variables
               force.in = NULL, force.out = NULL,
               method = "exhaustive")
summary_best_subset <- summary(Best_Subset)
as.data.frame(summary_best_subset$outmat)
```

```r
which.max(summary_best_subset$adjr2)
```

```r
summary_best_subset$which[8,]
```

```r
best.model <- lm(target_q_mean ~ mean_potential_evaporation + mean_snow_density + mean_snow_depth + mean_snowwater_equiv + 
                 mean_snowfall+ mean_snowmelt+mean_temp + mean_total_precip, data = df)
summary(best.model)
```

```r
boruta_output <- Boruta(target_q_mean ~ ., data=na.omit(df), doTrace=0) 
```

```r
names(boruta_output)
```

```r
roughFixMod <- TentativeRoughFix(boruta_output)
boruta_signif <- getSelectedAttributes(roughFixMod)
print(boruta_signif)
```

```r
# Variable Importance Scores
imps <- attStats(roughFixMod)
imps2 = imps[imps$decision != 'Rejected', c('meanImp', 'decision')]
head(imps2[order(-imps2$meanImp), ])  # descending sort
```

```r
# Plot variable importance
plot(boruta_output, cex.axis=.7, las=2, xlab="", main="Variable Importance")  
```

```r
library(caret)
set.seed(100)
rPartMod <- train(target_q_mean ~ ., data=df, method="rpart")
rpartImp <- varImp(rPartMod)
print(rpartImp)
```
